```python
# get image urls
from tensorflow.keras.utils import load_img, img_to_array
image_urls = []
for i in data1:
    for j in i['products']:

        image_urls.append(j['imageUrl'])


#add an  http: // to the beginning of each url
image_urls = [r'http://' + i for i in image_urls]
 
```

```python
import requests
import os

# Make a directory to store the images
os.makedirs("images", exist_ok=True)

# Define a list of image URLs
# image_urls = [
#     'http://images.asos-media.com/products/akjaerbede-gray-round-cat-eye-sunglasses-in-black/201748665-1-black',
#     'http://images.asos-media.com/products/akjaerbede-marvin-round-sunglasses-in-champagne/201748564-1-beige',
#     'http://images.asos-media.com/products/akjaerbede-marvin-round-sunglasses-in-yellow-transparent/201748604-1-yellow',
#     'http://images.asos-media.com/products/akjaerbede-bror-rectangle-sunglasses-in-petroleum-transparent/201748696-1-blue',
#     'http://images.asos-media.com/products/akjaerbede-winnie-round-sunglasses-in-yellow-transparent/201748712-1-yellow',
#     'http://images.asos-media.com/products/akjaerbede-marvin-round-sunglasses-in-demi-tortoise/201748718-1-brown',
# ]

# Loop through the list of URLs and download each image
for i, url in enumerate(image_urls):
    response = requests.get(url)
    open(f"images/image{i}.jpg", "wb").write(response.content)
```
